```python
import tensorflow as tf
import numpy as np
```

```python
x = np.array([-40, -10, 0, 8, 15, 22],dtype=float)
y = np.array([-40, 14, 32, 46, 59, 72],dtype=float)
```

Vertical stack of nodes

```python
layer=tf.keras.layers.Dense(units=1, input_shape=[1])
```

Horizontal stack of layers

```python
network = tf.keras.Sequential([layer])
```

```python
network.compile(loss="mean_squared_error",optimizer=tf.keras.optimizers.Adam(learning_rate=0.1))
```

```python
network.summary()
```

```python
layer.get_weights()
```

```python
history = network.fit(x,y,epochs=500,verbose=True)
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Epoch Number")
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

```python
layer.get_weights()
```

```python
network.predict([38])
```

Multilayer

```python
layer0=tf.keras.layers.Dense(units=4, input_shape=[1])
layer1=tf.keras.layers.Dense(units=4)
layer2=tf.keras.layers.Dense(units=1)

network = tf.keras.Sequential([layer0,layer1,layer2])

network.compile(loss="mean_squared_error",optimizer=tf.keras.optimizers.Adam(learning_rate=0.1))
```

```python
network.summary()
```

```python
layer1.get_weights()
```

```python
history = network.fit(x,y,epochs=500,verbose=True)
```

```python
network.predict([38])
```
